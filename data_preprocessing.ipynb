```python
query = """
SELECT
    COUNT(*) AS total_rows,
    MIN(creation_date) AS min_creation_date,
    MAX(creation_date) AS max_creation_date
FROM
    `bigquery-public-data.stackoverflow.posts_questions`
WHERE
    post_type_id = 1 -- Убеждаемся, что считаем только вопросы
"""
stats_df = bq_assistant.query_to_pandas_safe(query, max_gb_scanned=10) # 10 GB - разумный лимит

total_rows = stats_df['total_rows'].iloc[0]
min_date = stats_df['min_creation_date'].iloc[0]
max_date = stats_df['max_creation_date'].iloc[0]

print(f"\nОбщее количество вопросов: {total_rows}")
print(f"Минимальная дата создания вопроса: {min_date}")
print(f"Максимальная дата создания вопроса: {max_date}")
```

```python
from datetime import datetime, timedelta

min_time = datetime(year=2022, month=1, day=1)
max_time = datetime(year=2022, month=6, day=1)
query = f"""
SELECT
    COUNT(*) AS num_questions
FROM
    `bigquery-public-data.stackoverflow.posts_questions`
WHERE
    creation_date >= '{min_time.strftime('%Y-%m-%d %H:%M:%S UTC')}' 
    AND creation_date <= '{max_time.strftime('%Y-%m-%d %H:%M:%S UTC')}'
    AND post_type_id = 1 -- Убеждаемся, что считаем только вопросы
"""

stats_df = bq_assistant.query_to_pandas_safe(query, max_gb_scanned=10) 

num_questions = stats_df['num_questions'].iloc[0]

print(f"\nКоличество вопросов, созданных с {min_time.strftime('%Y-%m-%d')} по {max_time.strftime('%Y-%m-%d')}: {num_questions}")
```

```python
target_tags = ['python', 'machine-learning', 'deep-learning', 'tensorflow', 'pytorch']

tags_filter_conditions = []
for tag in target_tags:
    tags_filter_conditions.append(f"tags = '{tag}'")  # Для случая, когда это единственный тег
    tags_filter_conditions.append(f"tags LIKE '{tag}|%'") # Тег в начале
    tags_filter_conditions.append(f"tags LIKE '%|{tag}'") # Тег в конце
    tags_filter_conditions.append(f"tags LIKE '%|{tag}|%'") # Тег в середине
final_tags_filter = " OR ".join(tags_filter_conditions)

query = f"""
SELECT
    COUNT(*) AS num_questions_filtered
FROM
    `bigquery-public-data.stackoverflow.posts_questions`
WHERE
    creation_date >= '{min_time.strftime('%Y-%m-%d %H:%M:%S UTC')}' 
    AND creation_date <= '{max_time.strftime('%Y-%m-%d %H:%M:%S UTC')}'
    AND post_type_id = 1 -- Только вопросы
    AND ({final_tags_filter}) -- Добавляем условие по тегам
"""

stats_df = bq_assistant.query_to_pandas_safe(query, max_gb_scanned=10)

num_questions = stats_df['num_questions_filtered'].iloc[0]

print(f"\nКоличество вопросов с выбранными тегами: {num_questions}")
```

```python
import pandas as pd
import os 

query = f"""
SELECT
    id,
    title,
    body,
    view_count,
    tags,
    owner_user_id,
    creation_date,
    score,
    answer_count,
    comment_count
FROM
    `bigquery-public-data.stackoverflow.posts_questions`
WHERE
    creation_date >= '{min_time.strftime('%Y-%m-%d %H:%M:%S UTC')}' 
    AND creation_date <= '{max_time.strftime('%Y-%m-%d %H:%M:%S UTC')}'
    AND post_type_id = 1
    AND ({final_tags_filter})
ORDER BY
    creation_date DESC
"""

df = bq_assistant.query_to_pandas_safe(query, max_gb_scanned=40)
```

# Обработка и созранение датасета

```python
# --- Обработка и сохранение данных ---

# 1. Преобразуем колонку 'tags' из строки в список
# В BigQuery теги могут быть None (если пост без тегов), поэтому проверяем тип
df['tags'] = df['tags'].apply(lambda x: x.split('|') if isinstance(x, str) else [])

# 2. Очистка тела поста от HTML-тегов (Stack Overflow body содержит HTML)
from bs4 import BeautifulSoup
import re

def clean_html(text):
    if not isinstance(text, str): # Проверяем, что это строка
        return ""
    soup = BeautifulSoup(text, 'html.parser')
    # Заменяем <pre> и <code> блоки на их содержимое, окруженное маркерами
    for pre_tag in soup.find_all('pre'):
        pre_tag.replace_with(f"[CODE_START] {pre_tag.get_text()} [CODE_END]")
    for code_tag in soup.find_all('code'):
        # Если <code> внутри <pre>, он уже обработан. Иначе обрабатываем.
        if code_tag.find_parent('pre') is None:
            code_tag.replace_with(f"[INLINE_CODE_START] {code_tag.get_text()} [INLINE_CODE_END]")
    # удаляем разметку
    clean_text = soup.get_text()
    return clean_text

print("\nОчистка HTML из тела постов. Это может занять некоторое время...")
df['body_cleaned'] = df['body'].apply(clean_html)
print("Очистка завершена.")

# 3. Объединяем заголовок и очищенное тело поста в один текстовый признак
df['full_text'] = df['title'] + "\n" + df['body_cleaned']

# Выбираем колонки, которые нам нужны для дальнейшей работы (без исходного body и title)
# А также убеждаемся, что view_count - это число (может быть float в некоторых случаях)
df['view_count'] = df['view_count'].fillna(0).astype(int)
df['score'] = df['score'].fillna(0).astype(int)
df['answer_count'] = df['answer_count'].fillna(0).astype(int)
df['comment_count'] = df['comment_count'].fillna(0).astype(int)
```

```python
from langdetect import detect, DetectorFactory

# --- Установка seed для воспроизводимости ---
DetectorFactory.seed = 0

# --- Функция для безопасного определения языка ---
# langdetect может выбрасывать ошибки для очень коротких или некорректных текстов.
# Эта функция обертывает вызов detect() в try-except.
def safe_detect_lang(text):
    if not isinstance(text, str) or len(text.strip()) < 10:
        # Слишком короткие или нестроковые значения не могут быть надежно определены
        return 'unknown_or_too_short'
    try:
        # Пытаемся определить язык
        return detect(text)
    except Exception:
        # Если произошла ошибка (например, из-за непонятного текста), помечаем как 'error'
        return 'error'

# --- Применение определения языка к DataFrame ---
print("Начинается проверка языка текстов. Это может занять некоторое время...")

# Создаем новую колонку 'detected_language' и применяем функцию
df['detected_language'] = df['full_text'].apply(safe_detect_lang)

# --- Анализ и фильтрация результатов ---
print("\nРаспределение обнаруженных языков:")
language_counts = df['detected_language'].value_counts()
print(language_counts)

# Сохраняем исходное количество строк для сравнения
initial_rows_count = len(df)

# Фильтруем DataFrame, оставляя только посты, определенные как английские ('en')
# Используем .copy(), чтобы избежать предупреждений Pandas (SettingWithCopyWarning)
df_filtered_lang = df[df['detected_language'] == 'en'].copy()

print(f"\nФильтрация по языку завершена.")
print(f"Исходное количество постов: {initial_rows_count}")
print(f"Количество постов после фильтрации на английский язык: {len(df_filtered_lang)}")

# Теперь 'df_filtered_lang' содержит только англоязычные посты
# Вы можете продолжить работу с этим DataFrame
print("\nПервые 5 строк отфильтрованного DataFrame:")
print(df_filtered_lang.head())
```

```python
from collections import Counter

# Сначала собираем все теги из всех постов
all_tags = [tag for sublist in df_filtered_lang['tags'] for tag in sublist]
tag_counts = Counter(all_tags)

# Определяем порог (например, тег должен встречаться минимум 50 раз)
min_tag_frequency = 50
frequent_tags = {tag for tag, count in tag_counts.items() if count >= min_tag_frequency}

print(f"\nОбщее количество уникальных тегов: {len(tag_counts)}")
print(f"Количество тегов, встречающихся >= {min_tag_frequency} раз: {len(frequent_tags)}")

# Фильтруем колонку 'tags' в DataFrame, оставляя только частые теги
df_filtered_lang['tags_filtered'] = df_filtered_lang['tags'].apply(
    lambda tag_list: [tag for tag in tag_list if tag in frequent_tags]
)

# Теперь отфильтруем посты, у которых после этой операции не осталось ни одного тега
initial_rows_after_lang_filter = len(df_filtered_lang)
df_final = df_filtered_lang[df_filtered_lang['tags_filtered'].apply(len) > 0].copy()
print(f"Отфильтровано постов, у которых не осталось частых тегов: {len(df_final)} из {initial_rows_after_lang_filter} строк.")
```

# Создание графа

```python
import pandas as pd
import networkx as nx
from collections import Counter, defaultdict
import os
import itertools 

print(f"Количество постов с тегами для графа: {len(df_final)}")


# --- Шаг 1: Эффективный сбор данных о ребрах (совместных появлениях) ---
print("\nЭффективный сбор данных о совместных появлениях тегов...")

# Используем defaultdict для хранения весов ребер
# Ключ: кортеж (tag1, tag2), значение: вес
edge_weights = defaultdict(int)
all_unique_tags = set() # Для сбора всех уникальных тегов (узлов)

for index, row in df_final.iterrows():
    tags = row['tags_filtered']

    # Добавляем все теги поста в множество уникальных тегов
    all_unique_tags.update(tags)

    # Генерируем все уникальные пары тегов для текущего поста
    # sorted(tag) нужен, чтобы (tag1, tag2) и (tag2, tag1) рассматривались как одна и та же пара
    for tag1, tag2 in itertools.combinations(sorted(tags), 2):
        edge_weights[(tag1, tag2)] += 1

print(f"Сбор данных о ребрах завершен. Обнаружено {len(all_unique_tags)} уникальных тегов и {len(edge_weights)} уникальных пар тегов.")


# --- Шаг 2: Построение графа из собранных данных ---
print("\nПостроение графа networkx из собранных данных...")

G_tags = nx.Graph()

# Добавляем все уникальные теги как узлы
G_tags.add_nodes_from(list(all_unique_tags))

# Добавляем ребра с весами
# Преобразуем edge_weights в список кортежей (u, v, {'weight': w}) для add_weighted_edges_from
edges_to_add = [(tag1, tag2, weight) for (tag1, tag2), weight in edge_weights.items()]
G_tags.add_weighted_edges_from(edges_to_add)

print(f"Граф успешно построен. Количество узлов (тегов): {G_tags.number_of_nodes()}")
print(f"Количество ребер (связей между тегами): {G_tags.number_of_edges()}")
```

```python
# Пример: вывести несколько узлов и их соседей с весами 
print("\nПримеры узлов и их соседей:")
for i, node in enumerate(list(G_tags.nodes())[:5]): # Вывести только 5 примеров
    print(f"Тег: {node}")
    # Вывести 3 самых сильных связи для этого тега
    # Проверяем, что у узла есть соседи
    if G_tags[node]:
        sorted_neighbors = sorted(G_tags[node].items(), key=lambda item: item[1]['weight'], reverse=True)
        for neighbor, attributes in sorted_neighbors[:3]:
            print(f"  - Связан с {neighbor} (вес: {attributes['weight']})")
    else:
        print("  - Нет соседей (изолированный узел)")
```

```python
# --- Шаг 3: Сохранение графа ---
graph_file_path = 'so_graph.graphml'

print(f"\nСохранение графа в {graph_file_path}...")
nx.write_graphml(G_tags, graph_file_path)
print("Граф успешно сохранен!")
```
